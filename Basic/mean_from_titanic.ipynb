```python
dataset=pd.read_csv("E:/DataScience-Portfolio/CSV,XL/titanic.csv")
```

```python
sns.histplot(x="Age",data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,50)],[i for  i in range(0,50)],c="yellow",label="mean")
plt.plot([md for i in range(0,50)],[i for  i in range(0,50)],c="green",label="median")
plt.plot([mode for i in range(0,50)],[i for  i in range(0,50)],c="red",label="mode")
plt.legend()
plt.show()
```
